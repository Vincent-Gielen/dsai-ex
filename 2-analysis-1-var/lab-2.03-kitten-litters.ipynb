## Exercise 3 - Kitten litters

A vet keeps a record of the number of kittens in litters produced by cats in his practice. The code block below contains the data as a frequency table.

- First put the data into one list of individual observations: [1, 1, 2, 2, ... 5, 5, 6, 6, 6, 6, 7, 7, 8]  
- Visualise the data set using all suitable chart types that you know of.
- Calculate:
    - the median size of a litter
    - the mean size of a litter
    - the lower quartile
    - the upper quartile

```python
kittens = pd.DataFrame.from_dict({
    'Number in litter': [1, 2, 3,  4, 5, 6, 7, 8],
    'Frequency':        [2, 4, 7, 11, 8, 4, 2, 1]})
kittens
```

```python
kittensArray = []
# NumberInLitterArray = kittens['Number in litter']
# FrequencyArray = kittens.Frequency
# for x in range(len(FrequencyArray)):
#   for y in range(FrequencyArray[x]):
#     kittensArray.append(NumberInLitterArray[x])
kittensArray = np.repeat(kittens['Number in litter'], kittens['Frequency']).tolist()
kittensArray
```

```python
sns.boxplot(data=kittensArray)
```

```python
sns.displot(data=kittensArray, kde=True)
```

```python
sns.kdeplot(data=kittensArray)
```

```python
sns.violinplot(x=kittensArray)
```

```python
print(f"Median: {np.median(kittensArray)}")
print(f"Mean: {np.mean(kittensArray)}")
print(f"Lower quartile: {np.quantile(kittensArray, 0.25)}")
print(f"Upper quartile: {np.quantile(kittensArray, 0.75)}")
```
